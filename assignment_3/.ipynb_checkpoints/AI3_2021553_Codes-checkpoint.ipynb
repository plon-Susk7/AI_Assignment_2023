```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../wine/wine.data')

df.columns = ['class','alcohol','maclic acid','ash','alcalinity','mg','phenols','flavanoids','nf phenols','proanthocyanins','color intensity','hue','diluted wines','proline']
```

```python
df.dropna()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.metrics import accuracy_score


discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
data_discretized = pd.DataFrame(discretizer.fit_transform(df), columns=df.columns)

# Split the data into features and target variable
X = data_discretized.drop('class', axis=1)
y = data_discretized['class']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

valid = pd.concat([X_test, y_test], axis='columns')
dfnum = pd.concat([X_train, y_train], axis='columns')
dfnum
```

```python
import bnlearn as bn

# Structure learning
DAG = bn.structure_learning.fit(dfnum, methodtype='hc', bw_list_method='nodes', verbose=3)

# Plot
G = bn.plot(DAG)

# Parameter learning
model = bn.parameter_learning.fit(DAG, dfnum, methodtype='ml',verbose=3);
```

```python
def get_acc(model, df, col):
    # Get accuracy score by the model for the validation dataset df with target col
    print(df[col])
    pred = bn.predict(model, df, variables=[col])
    print(pred)
    acc = accuracy_score(df[col], pred[col])
    print('Accuracy -', acc)
    return acc

acc1 = get_acc(model, valid, 'class')
```
